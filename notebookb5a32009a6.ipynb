```python
import numpy as np
import pandas as pd
```

```python
tracks = pd.read_csv('/kaggle/input/-spotify-tracks-dataset/dataset.csv')
tracks.head()
```

```python
tracks = tracks.drop('Unnamed: 0', axis=1)
tracks.head()
```

```python
print(tracks.isnull().sum())
```

```python
tracks.dropna(inplace = True)
```

```python
tracks.dtypes
```

```python
tracks = tracks.drop(columns=["duration_ms"])
tracks.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = [ 
    "popularity", "danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "time_signature"
]

scaler = MinMaxScaler()
tracks[num_cols] = scaler.fit_transform(tracks[num_cols])
tracks.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

# Encoding categorical features
tracks["artists"] = encoder.fit_transform(tracks["artists"])
tracks["album_name"] = encoder.fit_transform(tracks["album_name"])
tracks["track_genre"] = encoder.fit_transform(tracks["track_genre"])

# Fixing explicit column (True/False â†’ 1/0)
tracks["explicit"] = tracks["explicit"].astype(int)
tracks.head()
```

```python
from sklearn.neighbors import NearestNeighbors

# Drop non-numeric columns
features = tracks.drop(columns=["track_id", "track_name"])

# Fit KNN Model
nn_model = NearestNeighbors(n_neighbors=10, metric="cosine", algorithm="auto")
nn_model.fit(features)

def get_similar_songs(track_index, n=10, artist_bias=0.1):
    # Ensure track_index is within range
    if track_index >= len(features) or track_index < 0:
        raise ValueError("Track index is out of bounds.")

    # Extract features of the given track and keep column names
    track_features = pd.DataFrame([features.iloc[track_index]], columns=features.columns)

    # Find nearest neighbors
    distances, indices = nn_model.kneighbors(track_features, n_neighbors=n+1)

    # Retrieve similar songs (excluding the input track itself)
    similar_songs = tracks.iloc[indices[0][1:]].copy()
    
    # Use encoded "artist" column instead of "artists"
    original_artist = tracks.iloc[track_index]["artists"]

    # Adjust distances for same artist (vectorized)
    artist_mask = (similar_songs["artists"] == original_artist).astype(float)
    similar_songs["adjusted_distance"] = distances[0][1:] - (artist_bias * artist_mask)

    # Sort by adjusted distance
    similar_songs = similar_songs.sort_values(by="adjusted_distance").head(n)

    return similar_songs[["track_id", "track_name"]]


# Example usage
similar_songs = get_similar_songs(81004, n=10, artist_bias=0.9)
print(similar_songs)
```
